Starting with Bivaroate Analysis:
1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='size')
```

2. Bar Plot(Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='pclass',y='fare',data=titanic,hue='sex')
```

3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='sex',y='age',data=titanic,hue='survived')
```

4. Dist Plot (Numerical - Categorical)

```python
sns.distplot(titanic['age'])
```

```python
sns.histplot(data=titanic, x='age', hue='survived', kde=True, element="step")
```

```python
sns.kdeplot(data=titanic, x='age', hue='survived', fill=True)
```

5. Heat Map (Categorical - Categorical)

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()*100
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

6. Cluster Map (categorical - Categorical)

```python
pd.crosstab(titanic['parch'],titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

7. Pair Plot (Nothing bu collection of scatter plots)

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

8. Line Plot (Numerical - Numerical) => use when x axis has time based

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
new
```

```python
sns.lineplot(x='year', y='passengers',data=new)
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
