COLUMN TRANSFORMER

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
X_train
```

1. Aam Zindagi

```python
# adding simple imputer to fever col
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])

# also the test data
X_test_fever = si.fit_transform(X_test[['fever']])

X_train_fever.shape
```

```python
# Ordinalencoding -> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])

# also the test data
X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```python
# OneHotEncoding -> gender,city
ohe = OneHotEncoder(drop='first')
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

# also the test data
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
# Force gender_city to match the 80 rows of age
X_train_gender_city_fixed = np.broadcast_to(X_train_gender_city, (X_train_age.shape[0], X_train_gender_city.shape[1] if X_train_gender_city.ndim > 1 else 1))
X_test_gender_city_fixed = np.broadcast_to(X_test_gender_city, (X_test_age.shape[0], X_test_gender_city.shape[1] if X_test_gender_city.ndim > 1 else 1))

# Now concatenate using your original style
X_train_transformed = np.column_stack((X_train_age, X_train_fever, X_train_gender_city_fixed, X_train_cough))
X_test_transformed = np.column_stack((X_test_age, X_test_fever, X_test_gender_city_fixed, X_test_cough))

print(X_train_transformed.shape)
```

2. Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```
