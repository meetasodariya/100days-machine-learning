```python
import pandas as pd
import requests
```

Just for ref: https://api.themoviedb.org/3/movie/top_rated?api_key=8d4d1c35a4adf840fa65da34280acc73&language=en-US&page=1

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8d4d1c35a4adf840fa65da34280acc73&language=en-US&page=1')
```

```python
response
```

```python
response.json()['results']
```

```python
pd.DataFrame(response.json()['results'])
```

```python
df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df
```

Doing everything above into the loop to convert the all data into the frame

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8d4d1c35a4adf840fa65da34280acc73&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = pd.concat([df, temp_df], ignore_index=True)
```

```python
df
```

```python
df.shape
```

```python
df.to_csv('movies.csv')
```
