```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
df.shape
```

```python
df.head()
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

```python
# --- 0. Sample Data Setup (For demonstration) ---
# Assuming 'df' exists with 'feature1', 'feature2', 'target'
np.random.seed(42)
X = np.random.uniform(-5, 5, 50)
Y = np.random.uniform(-5, 5, 50)
Z = 0.5 * X + 0.2 * Y + np.random.normal(0, 1, 50)
df = pd.DataFrame({'feature1': X, 'feature2': Y, 'target': Z})
lr = LinearRegression().fit(df[['feature1', 'feature2']], df['target'])

# --- 1. Define Range ---
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# --- 2. Create Grid ---
xGrid, yGrid = np.meshgrid(x, y)

# --- 3. Create 'final' BEFORE prediction ---
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T

# --- 4. Predict and Reshape ---
z_final = lr.predict(final).reshape(10, 10)

# --- 5. Plotting Together ---
# A. Create the scatter plot using Express
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target', opacity=0.7)

# B. Add the surface trace using Graph Objects
fig.add_trace(
    go.Surface(x=x, y=y, z=z_final, opacity=0.5, colorscale='Viridis')
)

# C. Show plot
fig.show()
```

Making our own Linear Regression Class

```python
class MeraLR:

    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        X_train = np.insert(X_train,0,1,axis=1)

        # calcuate the coeffs
        betas = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]

    def predict(self,X_test):
        y_pred = np.dot(X_test,self.coef_) + self.intercept_
        return y_pred
```

```python
lr = MeraLR()
```

```python
lr.fit(X_train,y_train)
```
